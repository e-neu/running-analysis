```python
%matplotlib inline

import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("../01_data/athlete_events.csv")
```

```python
running_event_list = ["Athletics Men's 100 metres", "Athletics Men's 200 metres", "Athletics Men's 400 metres", "Athletics Men's 800 metres", "Athletics Men's 1,500 metres", "Athletics Men's 5,000 metres", "Athletics Men's 10,000 metres", "Athletics Men's Marathon"]
```

```python
runner_data = data[(data['Sport'] == 'Athletics') & (data['Event'].str.contains('|'.join(running_event_list)))].copy()
```

```python
runner_data['BMI'] = runner_data['Weight']/(runner_data['Height']*runner_data['Height']/10000)
```

```python
g = sns.catplot(x='Event', y='Weight', order=running_event_list, kind="boxen", data=runner_data[runner_data['Medal'].notnull()])
g.set_xticklabels(rotation=90)
```

```python
g = sns.lmplot(x='Year', 
               y='Weight', 
               col='Event', 
               hue='Event', 
               data=runner_data[runner_data['Medal'].notnull()], 
               col_wrap=4,
               col_order=running_event_list,
              )
g.set_xticklabels(rotation=90)
```

```python
g = sns.lmplot(x='Year', 
               y='BMI', 
               col='Event', 
               hue='Event', 
               data=runner_data[runner_data['Medal'].notnull()], 
               col_wrap=4,
               col_order=running_event_list,
              )
g.set_xticklabels(rotation=90)
```
